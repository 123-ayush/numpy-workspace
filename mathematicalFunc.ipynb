Important Mathematical Functions

Aggregation
- sum( )
- prod( )
- min( ) -> min ele in arr
- max( ) -> max ele in arr
- mean( ) -> mean
- std( ) -> std dev
- var( ) -> variance(square of std dev)
- median( ) -> median
- argmin( ) -> return idx of min val
- argmax( ) -> return idx of max val

Power
- square()
- sqrt( )
- pow(a, b)

Log & Exponential
- log() -> base e
- log10()
- log2()
- exp()

Rounding
- round( ) -> rounding the float
- floor( ) -> nearest integer < val
- ceil( ) -> nearest integer just > val
- trunc( ) -> print but cutting all the decimal part

EXTRAS
- unique()
- sort( )
- abs( )
