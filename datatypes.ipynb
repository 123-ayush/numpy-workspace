NumPy Common Data Types:
- Integers: int32, int64
- Floating point nums: float32, float64
- Boolean: bool
- Complex nums: complex64, complex128
- String: S (byte-str) & U (unicode-str)
- Object: generic python objects - object

```python
import numpy as np

num = 45
print(num, type(num))

num = 45.34
print(num, type(num))

num = np.array([3 + 4j])
print(num, num.dtype)

num = True
print(num, type(num))

num = 'ayush'
print(num, type(num))

num = np.array([1, 2, 'prime', {1,2,3}, 3.14])
print(num, num.dtype)
```
